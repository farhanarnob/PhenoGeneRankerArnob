```python
g_e_corr_full_read_round_2 = pd.read_csv('/content/drive/MyDrive/myself/UNT/sem one/individual research/project/PhenoGeneRankerArnob/database/ADNI_Gene_Expression_Profile/g_e_corr_full_read_round_2.csv',index_col=0)
```

```python
g_e_corr_full_read_round_2_strong = g_e_corr_full_read_round_2[g_e_corr_full_read_round_2 >= 0.7]
```

```python
g_e_corr_full_read_round_2_strong_stack = g_e_corr_full_read_round_2_strong.stack().reset_index()
```

```python
g_e_corr_full_read_round_2_strong_stack = g_e_corr_full_read_round_2_strong_stack.rename(columns={'level_0': 'from','level_1': 'to',0:"weight"})
g_e_corr_full_read_round_2_strong_stack['from'] = g_e_corr_full_read_round_2_strong_stack['from'].astype(int)
g_e_corr_full_read_round_2_strong_stack['to'] = g_e_corr_full_read_round_2_strong_stack['to'].astype(int)
```

```python
g_e_corr_full_read_round_2_strong_stack.head(10)
```

```python
index_names = g_e_corr_full_read_round_2_strong_stack[ (g_e_corr_full_read_round_2_strong_stack['from'] == g_e_corr_full_read_round_2_strong_stack['to'])].index

index_names
```

```python
g_e_corr_full_read_round_2_strong_stack.drop(index_names, inplace = True)
```
